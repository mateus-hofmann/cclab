## 3. Retrieved Documents

**Web of Sience** 

```python
import pandas as pd
import re

# Adjust Pandas display options
pd.set_option('display.max_colwidth', 50)
pd.set_option('display.width', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.show_dimensions', False)

# Function to extract and clean the country from the reprint address
def extract_country_clean(address):
    if pd.isnull(address):
        return None
    # Assumes the country is at the end, after the last comma or semicolon
    parts = address.split(';')[-1]  # Splits by semicolon and takes the last part
    country = parts.split(',')[-1].strip()  # Splits by the last comma and removes extra spaces
    # Removes numbers and common US state abbreviations
    country_clean = re.sub(r'\b(\d+|[A-Z]{2})\b', '', country).strip()
    # Removes trailing dots
    country_clean = country_clean.rstrip('.')
    return country_clean

# Path to the Excel file
file_path = "wos_env.xls"

# Read the Excel file
df = pd.read_excel(file_path)

# Apply the function to extract and clean countries
df['Country'] = df['Reprint Addresses'].apply(extract_country_clean)

# Desired columns, now including 'Country'
selected_columns = ["Article Title", "Abstract", "Publication Year", "Document Type", "Author Full Names", "Source Title", "Country", "DOI"]

# Select the desired information and create a copy
filtered_df_wos = df[selected_columns].copy()

# Rename columns as per request
filtered_df_wos.columns = ["Title", "Abstract", "Year", "Type", "Authors", "Source", "Country", "DOI"]

# Print the number of articles found in the Excel file
print("Number of documents found:", len(filtered_df_wos))

# Display the filtered DataFrame in Jupyter Lab
display(filtered_df_wos)
```

**Scopus**

```python
import pandas as pd
import re

# Adjust Pandas display options
pd.set_option('display.max_colwidth', 50)
pd.set_option('display.width', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.show_dimensions', False)

# Function to clean the full names of authors, removing numbers within parentheses
def clean_full_author_names(author_names):
    # Remove numbers and parentheses using a regular expression
    clean_names = re.sub(r'\(\d+\)', '', author_names)
    return clean_names.strip()

# Function to clean and extract the country from the correspondence address
def extract_clean_country(address):
    if pd.isnull(address) or not isinstance(address, str):
        return None
    
    # Split the string to remove emails and other irrelevant information
    address_parts = re.split(r'\s*[\w\.-]+@[\w\.-]+\.\w+\s*', address)
    address_without_email = address_parts[0] if address_parts else address
    
    # Split by comma to find the possible country, which is usually the last part
    address_segments = [segment.strip() for segment in address_without_email.split(',')]
    possible_country = address_segments[-1]
    
    # Clean the string to remove any unnecessary text like "email" or digits
    country = re.sub(r'\b(email|Email)\b', '', possible_country)  # Remove the word "email"
    country = re.sub(r'[0-9]+', '', country).strip()  # Remove digits
    country = re.sub(r'\([^)]*\)', '', country)  # Remove text in parentheses
    country = re.sub(r'[^\w\s]', '', country)  # Remove special characters
    
    # Some entries might have additional parts after the country; we try to detect this and remove it
    # Split by space and assume the country is the first word
    country_parts = country.split()
    clean_country = country_parts[0] if country_parts else country
    
    return clean_country

# Path to the Scopus Excel file
scopus_file_path = "scopus_env.xlsx"  # Update with the correct file path

# Read the Scopus Excel file
scopus_df = pd.read_excel(scopus_file_path)

# Clean the "Author full names" column
scopus_df['Author full names'] = scopus_df['Author full names'].apply(clean_full_author_names)

# Extract and clean the "Country" from the "Correspondence Address"
scopus_df['Country'] = scopus_df['Correspondence Address'].apply(extract_clean_country)

# Desired columns, now including 'Country' and cleaned 'Author full names'
selected_columns_scopus = ["Title", "Abstract", "Year", "Document Type", "Author full names", "Source title", "Country", "DOI"]

# Select the desired information and create a copy
filtered_df_scopus = scopus_df[selected_columns_scopus].copy()

# Rename columns as per request
filtered_df_scopus.columns = ["Title", "Abstract", "Year", "Type", "Authors", "Source", "Country", "DOI"]

# Print the number of articles found in the Excel file
print("Number of documents found:", len(filtered_df_scopus))

# Display the filtered DataFrame in Jupyter Lab
display(filtered_df_scopus)
```

## 4. Concatenation and De-Duplication

```python
# Concatenate the two DataFrames
combined_df = pd.concat([filtered_df_wos, filtered_df_scopus], ignore_index=True)

# Function to normalize strings: remove extra white spaces and convert to lowercase
def normalize_string(s):
    return s.strip().lower() if isinstance(s, str) else s

# Normalize relevant data
combined_df['Title'] = combined_df['Title'].apply(normalize_string)
combined_df['DOI'] = combined_df['DOI'].apply(normalize_string)
# combined_df['Year'] = combined_df['Year'].apply(str)  # Convert Year to string for consistency

# Remove duplicates based on 'Title', 'DOI', and 'Year'
unique_articles_df = combined_df.drop_duplicates(subset=['Title', 'DOI', 'Year'])

# Reset the index of the resulting DataFrame to reflect the removal of duplicates
unique_articles_df.reset_index(drop=True, inplace=True)

# Display the number of unique documents found
print("Number of unique documents found:", len(unique_articles_df))

# Display the DataFrame of unique articles
display(unique_articles_df)
```

```python
# Exporting DataFrame to Excel
output_file_path = "df_env.xlsx"
unique_articles_df.to_excel(output_file_path, index=False)
```
