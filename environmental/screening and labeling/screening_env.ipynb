## 3. Screening

### 3.1. Initializing

```python
import os
import pandas as pd
from IPython.display import display

# Path to the Excel file
file_path = "df_env.xlsx"

# Read the Excel file
df = pd.read_excel(file_path)

# Define your categories, subcategories, and subsubcategories
subcategories = [
    'Production_Material', 'Production_Manufacturing',
    'Construction_Installation', 'Construction_Traffic',
    'Use_Use/in', 'Use_Maintenance/Repair', 'Use_Replacement', 'Use_Operational',
    'EoL_Landfill', 'EoL_Incineration', 'EoL_Reuse', 'EoL_Recycling'
]

# Initialize the category columns with -1, indicating the articles have not yet been categorized
for subcat in subcategories:
    df[subcat] = -1

# Display the updated DataFrame
display(df)
```

```python
# Exporting DataFrame to Excel
output_file_path = "init_env.xlsx"
df.to_excel(output_file_path, index=False)
```
